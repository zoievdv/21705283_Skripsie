```python
df = pd.read_csv('data/doc_info_sentiment.csv', on_bad_lines='warn')
df
```

```python
def get_timeframe(topic):
    try:
        timeframe = topic.split('.')[0]
        return int(timeframe)
    except:
        return None

def get_topic(topic):
    try:
        return topic.split('.')[1]
    except:
        return None

df['timeframe'] = df['Topic_'].apply(get_timeframe)
df['topic'] = df['Topic_'].apply(get_topic)
df
```

```python
import matplotlib.pyplot as plt

df_grouped = df.groupby('timeframe').size().reset_index()
# df_grouped.plot()
df_grouped.columns = ['timeframe', 'count']
df_grouped
```

```python
# group by timeframe
# get count and average 'anger

df.groupby('timeframe').agg({'anger': ['count', 'mean']}).reset_index()
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
source = ColumnDataSource(df_grouped)


f = figure()

f.line(x='timeframe', y='count', source=source)

show(f)
```
